```python
df = df.drop_duplicates()       #The model learns from patterns, not from seeing the same row multiple times.
df.shape
```

```python
df.head()
```

```python
print(df['sex'].unique())
print(df['smoker'].unique())
print(df['region'].unique())
```

```python
df = pd.get_dummies(df, columns=['region'], drop_first=True, dtype=int)     #one hot encoding
df.head()
```

```python
df['sex'] = df['sex'].map({'female':0, 'male':1})
df['smoker'] = df['smoker'].map({'yes':1, 'no':0})

df.head()
```

```python
def bmi_category(bmi):
    if bmi<18.5:
        return 'underweight'
    elif 18.5 <= bmi <25:
        return 'normal'
    elif 25<= bmi < 30:
        return 'overweight'
    else:
        return 'Obese'
    
df['bmi_category'] = df['bmi'].apply(bmi_category)
df.head()
```

```python
#hot encoding
df = pd.get_dummies(df,columns=['bmi_category'],drop_first=True, dtype=int)

#interaction term : smoker x bmi
df['smoker_bmi'] = df['smoker'] * df['bmi']
df.head()
```

```python
df.to_csv('../data/cleaned_data.csv', index=False)
```
