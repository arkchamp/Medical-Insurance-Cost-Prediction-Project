```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['figure.figsize'] = (8,5)
```

```python
df = pd.read_csv('../data/cleaned_data.csv')
dfRaw = pd.read_csv('../data/medical_insurance.csv')
df.head()
```

```python
# Univariate Analysis
```

```python
# Q1: Distribution of medical insurance charges
sns.histplot(df['charges'], kde=True, bins=30, color='skyblue')
plt.title('Distb. of Medical Insurance Charges')
plt.xlabel('Charges')
plt.ylabel('Count')
plt.show()
```

```python
# Q2: Age distribution of the individuals
sns.histplot(df['age'], kde=True, bins=20, color='pink')
plt.title('age distb. of the indiviudals')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# Q3: How many people are smokers vs non-smokers
plt.figure(figsize=(6,4))
sns.countplot(x='smoker', data=df, palette='pastel')
plt.title('Smokers vs Non-Smokers')
plt.xlabel('Smoker (1 = Yes, 0 = No)')
plt.ylabel('Count')
plt.show()
```

```python
# Q4: What is the average BMI in the dataset

avg_bmi = df['bmi'].mean()
print(f'Average bmi: {avg_bmi:.2f}')

sns.histplot(df['bmi'], kde=True, bins =20, color='orange')
plt.axvline(avg_bmi, color = 'blue', linestyle = '--', label=f'Average BMI = {avg_bmi:.2f}')
plt.title('BMI Distb.')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#Q5: Which regions have the most number of policyholders

sns.countplot(data=dfRaw, x='region',
              order=dfRaw['region'].value_counts().index, palette='Set2')
plt.title('No. of Policyholder by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
#Bivariate Analysis
df.head()
```

```python
#Q1: How do charges vary with age

sns.scatterplot(x='age',y='charges',data=df, alpha=0.6)
plt.title('Charges VS Age')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Q2: Is there a difference in average charges between smokers

plt.figure(figsize=(6,4))
sns.barplot(data=df, x='smoker', y='charges', palette='Set2', ci=None) #ci None to remove Black line of confidence
plt.title('Avg Charges: Smokers vs Non-Smokers')
plt.xlabel('Smoker[1=Yes | 0=No]')
plt.ylabel('Charges')
plt.show()
```

```python
# Q3: Does BMI impact insurance charges?

sns.scatterplot(data=df,x='bmi', y='charges',alpha=0.6)
plt.title('BMI VS Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.show()
```

```python
# Q4: Do men or women pay more on average?

sns.barplot(x='sex', y='charges', data=df, ci=None, palette='pastel')
plt.title('Average Charges: Men vs Women')
plt.xlabel('Sex [1=Male | 0= Female]')
plt.ylabel('Average Charges')
plt.show()
```

```python
# Q5: Is there a correlation between the number of children and the insurance charges?

sns.barplot(data=df, x='children', y='charges', ci=None, palette='muted')
plt.title('Average Charges by no. of Children')
plt.xlabel('#Children')
plt.ylabel('Avg Charges')
plt.show()
```

```python
#Multivariate Analysis
df.head()
```

```python
# Q1: How does smoking status combined with age affect medical charges?

sns.scatterplot(x='age', y='charges', hue='smoker', data=df, alpha=0.6, palette='Set1')
plt.title('Charges vs Age by Smoking Status')
plt.xlabel('Age')
plt.ylabel('Charges')

plt.legend(title='Smoker', labels=['Yes', 'No'])
plt.show()
```

```python
#Q2: What is the impact of gender and region on charges for smokers?

dfSmokers = dfRaw[dfRaw['smoker'] == 'yes']

sns.barplot(data=dfRaw, x='sex', y='charges', hue='region',ci=None, palette='Set2')
plt.title('Average Charges by Gender and Region(Smokers Only)')
plt.xlabel('Sex')
plt.ylabel('Avg Charges')
plt.show()
```

```python
# Q3:How do age, BMI, and smoking status together affect insurance cost?

from matplotlib.lines import Line2D

sns.scatterplot(data=df, x='age',y='bmi',
                size='charges', hue='smoker',
                palette={1:'red', 0:'blue'},alpha=0.6)
plt.title('Age, BMI and Smoking status impact on Charges')

# Custom legend with circle markers
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Yes', markerfacecolor='red', markersize=8),
    Line2D([0], [0], marker='o', color='w', label='No', markerfacecolor='blue', markersize=6)
]

plt.legend(title='Smoker',handles=legend_elements)
plt.show()
```

```python
# Do obese smokers (BMI > 30) pay significantly higher than non-obese non-smokers?

# dfCompare=
dfCompare = df[((df['smoker_bmi'] >= 30) | (df['smoker_bmi']==0))]

dfCompare['os_nons'] = np.where(dfCompare['smoker_bmi'] >=30,'Obese Smoker', 'Non-Obese Non-Smoker')

#plot
sns.barplot(data=dfCompare, x='os_nons',y='charges', ci=None, palette='Set1')
plt.title('Average Charges: Obese Smokers vs Non-Obese Non-Smokers')
plt.xlabel('Group')
plt.ylabel('Average Charges')
plt.show()

print(dfCompare.groupby('os_nons')['charges'].mean())
```

```python
#Q1: What is the correlation between numeric features like age, BMI, number of children, and charges?

numeric_cols = ['age', 'bmi', 'children', 'charges']

corr_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numeric Features')
plt.show()
```

```python
#Q2: Which features have the strongest correlation with the target variable (charges)?

corr_with_charges = df.corr(numeric_only=True)['charges'].sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=corr_with_charges.index, y=corr_with_charges.values, palette='viridis')
plt.title('Correlation of Features with Charges')
plt.xlabel('')
plt.ylabel('Correlatoin coefficiant')
plt.xticks(rotation=90)
plt.show()
```
